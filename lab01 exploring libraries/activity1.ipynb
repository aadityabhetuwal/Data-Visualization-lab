```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
# a simple bar graph in matplotlib
N = 5

xvals = np.arange(1, N + 1)

yvals = np.random.randn(N) + 5
colors = ['r', 'g', 'b', 'y', 'k']

plt.bar(xvals, yvals, color=colors)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Bar plot in matplotlib")

plt.savefig("barChar.png")
```

```python
Image("barChar.png")
```

```python
# pie chart in matplotlib
N = 7

random_data = [np.random.randint(1, 100) for i in range(N)]
explodeRad = [0,] * 7
explodeRad[1] = .2

plt.pie(random_data, explode = explodeRad, autopct= "%.2f" )
plt.title("Pie chart in matplotlib")
plt.savefig("pieChart.png")
```
